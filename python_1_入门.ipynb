```python
# 函数
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# 类
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def passed(self):
        return self.score >= 60

# 异常
try:
    x = 10 // 2
except ZeroDivisionError:
    x = None

print(gcd(84, 18))
print(Student('Bob', 76).passed(), x)
```

#### 1.1 输入输出与类型转换
- 单个输入：`n = int(input())`
- 多个输入：`a, b = map(int, input().split())`
- 一行数组：`arr = list(map(int, input().split()))`
- 字符串输入：`s = input().strip()`

```python
import math

raw1 = '12'
raw2 = '3 5'
raw3 = '1 2 3 4 5'

n = int(raw1)
a, b = map(int, raw2.split())
arr = list(map(int, raw3.split()))

print('n =', n)
print('a+b =', a + b)
print('arr =', arr)
print('格式化输出: %.2f' % (math.pi))
print(f'格式化输出: {math.pi:.3f}')
```

#### 1.2 数组（List）与常用内置函数
重点：`len/sum/min/max/sorted/reversed/enumerate/zip`，以及 `append/pop/insert/remove`。

```python
arr = [5, 1, 3, 2, 4]
arr.append(6)
arr.insert(0, 9)
arr.remove(3)
x = arr.pop()

print('arr:', arr, 'pop:', x)
print('len/sum/min/max:', len(arr), sum(arr), min(arr), max(arr))
print('sorted:', sorted(arr))
print('reversed:', list(reversed(arr)))
print('enumerate:', [(i, v) for i, v in enumerate(arr)])
print('zip:', list(zip([1,2,3], ['a','b','c'])))
```

```python
# 字典、集合、字符串常用操作
d = {'alice': 95, 'bob': 88}
d['cindy'] = 91
d['bob'] = 90

st = set([1, 2, 2, 3, 4])
st.add(5)

s = 'python,algorithm'
print('dict keys:', list(d.keys()))
print('dict items:', list(d.items()))
print('set:', st)
print('split:', s.split(','))
print('upper:', s.upper())
print('replace:', s.replace('python', 'PY'))
```

```python
# 分支与循环：if/for/while/break/continue
nums = [1, 2, 3, 4, 5, 6]
odd_sum = 0
for x in nums:
    if x % 2 == 0:
        continue
    odd_sum += x

cnt = 0
cur = 1
while True:
    cnt += cur
    cur += 1
    if cnt >= 10:
        break

if odd_sum > 5:
    flag = 'ok'
else:
    flag = 'small'

print('odd_sum:', odd_sum, 'cnt:', cnt, 'flag:', flag)
```

```python
# 函数：默认参数、可变参数、lambda、排序key
def power(x, p=2):
    return x ** p

def add_all(*args):
    return sum(args)

pairs = [('a', 3), ('b', 1), ('c', 2)]
pairs_sorted = sorted(pairs, key=lambda x: x[1])

print(power(5), power(2, 3))
print(add_all(1, 2, 3, 4))
print(pairs_sorted)
```

```python
# 基础类定义与使用
class CounterBox:
    def __init__(self, start=0):
        self.value = start

    def inc(self, step=1):
        self.value += step

    def get(self):
        return self.value

box = CounterBox(10)
box.inc()
box.inc(5)
print('CounterBox:', box.get())
```
